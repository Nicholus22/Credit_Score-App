```python
import numpy as np
import pandas as pd

np.random.seed(42)
n = 3000

df = pd.DataFrame({
    "age": np.random.randint(21, 65, n),
    "monthly_income": np.random.randint(3000, 70000, n),
    "employment_years": np.random.randint(0, 35, n),
    "credit_history_years": np.random.randint(0, 25, n),
    "existing_loans": np.random.randint(0, 6, n),
    "total_debt": np.random.randint(0, 500000, n),
    "late_payments": np.random.randint(0, 12, n),
    "credit_utilization": np.round(np.random.uniform(0.05, 1.0, n), 2)
})
```

```python
df["default"] = (
    (df["late_payments"] >= 4) |
    (df["credit_utilization"] > 0.85) |
    (df["total_debt"] > df["monthly_income"] * 15)
).astype(int)

df.head()
```

```python
df.to_csv("synthetic_credit_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("synthetic_credit_data.csv")
df.info()
```

```python
#Distribution Checks

sns.countplot(x="default", data=df)
plt.show()
```

```python
sns.boxplot(x="default", y="monthly_income", data=df)
plt.show()
```

```python
#Correlation Matrix

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
#Derived Features

df["debt_to_income"] = df["total_debt"] / (df["monthly_income"] * 12)
df["loan_per_income"] = df["existing_loans"] / (df["monthly_income"] + 1)
```

```python
features = [
    "age",
    "monthly_income",
    "employment_years",
    "credit_history_years",
    "existing_loans",
    "total_debt",
    "late_payments",
    "credit_utilization",
    "debt_to_income",
    "loan_per_income"
]

X = df[features]
y = df["default"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=2000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

print(classification_report(y_test, model.predict(X_test)))
print("ROC AUC:", roc_auc_score(y_test, model.predict_proba(X_test)[:,1]))
```

```python
import joblib

joblib.dump(model, "credit_risk_model.pkl")
```

```python
import sqlite3

conn = sqlite3.connect("database/credit_risk.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS applicants (
    applicant_id INTEGER PRIMARY KEY AUTOINCREMENT,
    age INTEGER,
    monthly_income REAL,
    employment_years INTEGER,
    credit_history_years INTEGER,
    existing_loans INTEGER,
    total_debt REAL,
    late_payments INTEGER,
    credit_utilization REAL,
    debt_to_income REAL,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS credit_decisions (
    decision_id INTEGER PRIMARY KEY AUTOINCREMENT,
    applicant_id INTEGER,
    risk_probability REAL,
    risk_level TEXT,
    decision TEXT,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    FOREIGN KEY(applicant_id) REFERENCES applicants(applicant_id)
)
""")

conn.commit()
conn.close()
print("Database created successfully!")
```

```python
def risk_level(prob):
    if prob < 0.3:
        return "Low"
    elif prob < 0.6:
        return "Medium"
    else:
        return "High"

def credit_decision(prob):
    return "Approved" if prob < 0.5 else "Declined"
```
